# Year wise comparison of a particular State(Delhi)

```python
#importing the important libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#loading the dataset
acc= pd.read_csv('F:/Study/data/acc.csv',header=0,index_col=None)

#looking at the first 5 data
print(acc.shape)
acc.head()

```

1.Making two lists, in Acc we store the no. of accidents for diffent years of Delhi and in State we store the years corresponding to Acc.
2.Start the plot

```python
#Making list
Acc=[acc.Road_Accidents_2013[33],acc.Road_Accidents_2014[33],acc.Road_Accidents_2015[33],acc.Road_Accidents_2016[33]]
State=['2013','2014','2015','2016']

#making the grid for better visualization and understanding
plt.grid(color="green")

#naming the x and the y coordinate
plt.xlabel("Year")
plt.ylabel("Accidents")

#putting the title
plt.title("Delhi")

#bar plot
plt.bar(State,Acc)

#seeing the plot
plt.show()
```
