```python
#importing the important libraries 
import pandas as pd
import numpy as np

#loading the dataset
acc= pd.read_csv('F:/Study/data/acc.csv',header=0,index_col=None)

#looking at the shape of the data and first 5 data
print(acc.shape)
acc.head()

```

For number of accidents in the year 2013--
Since, the shape of the data is 37 and the last column gives the total amount of accidents so we won't use that here.
1.We assign a value a=0, this will work as a count variable in the future
2.Series of if else statement, this was done because we are plotting different plots to reduce the hapazardness
3.We make two lists and we append the values of the States and the Number of Accidents one by one
4.Finally, we plot the bar plot from the library matplotlib.pyplot

```python
a=0
while(a<36):
    if a==0:
        i=0
        j=11
    elif a==11:
        i=11
        j=23
    elif a==23:
        i=23
        j=33
    elif a==33:
        i=33
        j=36

    Acc=[]
    State=[]
    for i in range(i,j,1):
        State.append(acc.States[i])
        Acc.append(acc.Road_Accidents_2013[i])
        a+=1
    #importing the library to plot the data-- matplotlib.pyplot
    import matplotlib.pyplot as plt
    #to alter the size of the graph
    f,ax = plt.subplots(figsize=(20,5))
    #Naming the x and the y coordinate 
    plt.xlabel('State') 
    plt.ylabel("No. of Road Accidents")
    #title
    plt.title("Accidents 2013")
    #bar plot
    plt.bar(State,Acc)
    #seeing the plot
    plt.show()


```
